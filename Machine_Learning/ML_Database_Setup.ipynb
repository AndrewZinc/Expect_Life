```python
def bucket_le(le):
    if(le <= 45):
        le = 45
    elif(le > 45 and le <= 50):
        le = 50
    elif (le > 50 and le <= 55):
        le = 55
    elif (le > 55 and le <= 60):
        le = 60
    elif (le > 60 and le <= 65):
        le = 65
    elif (le > 65 and le <= 70):
        le = 70
    elif (le > 70 and le <= 75):
        le = 75
    elif (le > 75 and le <= 80):
        le = 80
    elif (le > 80 and le <= 85):
        le = 85
    return le
```

```python
ml_df['Life Expectancy']= ml_df['Life Expectancy'].apply(bucket_le)
```
