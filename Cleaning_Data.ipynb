```python
health_df["Series Name"] = health_df["Series Name"].to_string(header = False, index = False)
health_df["Series Name"]
```

```python
def parse_spending(item):
    if (item == 'Domestic general government health expenditure per capita (current US$)'):
        item = 'govt expenditure per capita in USD'
    elif (item == 'Domestic general government health expenditure (% of general government expenditure)'):
        item = 'govt health expenditure as % of total'
    elif (item == 'Domestic general government health expenditure (% of GDP)'):
        item = 'govt as % GDP'
    elif (item == 'Domestic general government health expenditure (% of current health expenditure)'):
        item = 'govt as % total health'
    elif (item == 'Domestic private health expenditure (% of current health expenditure)'):
        item = 'private as % total health'
    elif (item == 'Domestic private health expenditure per capita (current US$)'):
        item = 'private per capita in USD'
    elif (item == 'Domestic private health expenditure per capita, PPP (current international $)'):
        item = 'private in international $'
    return item
```
