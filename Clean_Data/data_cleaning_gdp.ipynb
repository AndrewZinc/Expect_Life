```python
# Import our dependencies
import pandas as pd

#  Import and read the charity_data.csv
systems_df = pd.read_csv("country_social_security_systems.csv")
systems_df
```

```python
gdp_df = pd.read_csv('../Raw_Data/GDP_en_csv_v2_4701247.csv')
gdp_df = gdp_df.drop(['Country Code', 'Indicator Code'], 1)
gdp_df.dtypes
```

```python
gdp_countries = gdp_df['Country Name'].tolist()
gdp_countries
```

```python
countries_df = pd.read_csv('../Clean_Data/country_profile_urls.csv')
countries_df = countries_df.drop(['profile_url'], 1)
```

```python
country_list = countries_df['country'].tolist()
country_list
```

```python
# clean the GDP data based on the SS country list
no_match = []

for country in gdp_countries:
    if country in country_list:
        continue
    else:
        no_match.append(country)
```

```python
no_match
```

```python
gdp_clean_df = gdp_df.drop(gdp_df[gdp_df['Country Name'].isin(no_match)].index.tolist())
gdp_clean_df
```

```python
gdp_clean_df = gdp_clean_df.drop(gdp_clean_df.columns[gdp_clean_df.columns.str.contains('^1')], axis=1)
gdp_clean_df = gdp_clean_df.drop(['Unnamed: 66'], 1)
```

```python
gdp_clean_df.to_csv('gdp_clean.csv', index=False)
```
