```python
# Import our dependencies
import pandas as pd
```

```python
#  Import and read the input csv
dps_df = pd.read_csv('../Cluster_Analysis_Data/daily-protein-supply-animal-veg.csv')
dps_df
```

```python
# Drop the udpated file index column
dps_df = dps_df.drop(['Unnamed: 0'], 1)
```

```python
dps_df
```

```python
dps_df = dps_df[dps_df['Year']==2013]
dps_df
```

```python
# Open the master list of countries
countries_df = pd.read_csv('../Clean_Data/country_profile_urls.csv')
countries_df = countries_df.drop(['profile_url'], 1)
```

```python
country_list = countries_df['country'].tolist()
country_list
```

```python
# Create a list of Food Supply countries
ps_countries = dps_df['Entity'].tolist()
ps_countries
```

```python
# clean the FS data based on the SS country list
no_match = []

for country in ps_countries:
    if country in country_list:
        continue
    else:
        no_match.append(country)
```

```python
# Likely need to do manual country name clean-up for accented characters and abbreviations.
no_match
```

```python
dps_clean_df = dps_df.drop(dps_df[dps_df['Entity'].isin(no_match)].index.tolist())
dps_clean_df
```

```python
dps_clean_df = dps_df.dropna()
dps_clean_df
```
