```python
# Import our dependencies
import pandas as pd
```

```python
#  Import and read the input csv
at_df = pd.read_csv("../Cluster_Analysis_Data/alcohol_tobacco.csv")
at_df = at_df.dropna()
at_df
```

```python
tot_alcohol = []
tobacco_use =[]
```

```python
for i in at_df['Series Name'].index:
    if i % 2 == 0:
        tot_alcohol.append(at_df['2018'][i])
tot_alcohol
```

```python
for i in at_df['Series Name'].index:
    if i % 2 == 1:
        tobacco_use.append(at_df['2018'][i])
tobacco_use
```

```python
at_df = at_df.drop_duplicates(subset=['Country Name'])
at_df
```

```python
# Create a list of Alcohol consumption / Tobacco use countries
at_countries = at_df['Country Name'].tolist()
at_countries
```

```python
# Build a new DataFrame from the component lists

at_use_df = pd.DataFrame()
at_use_df['country'] = at_countries
at_use_df['tot_alcohol_consumption'] = tot_alcohol
at_use_df['tobacco_use_%'] = tobacco_use

at_use_df
```

```python
# Open the master list of countries
countries_df = pd.read_csv('../master_country_list/country_profile_urls.csv')
#countries_df = countries_df.drop(['profile_url'], 1)
```

```python
country_list = countries_df['country'].tolist()
country_list
```

```python
# clean the AT data based on the SS country list
no_match = []

for country in at_countries:
    if country in country_list:
        continue
    else:
        no_match.append(country)
```

```python
# Likely need to do manual country name clean-up for accented characters and abbreviations.
no_match
```

```python
# Write the AT data to CSV to adjust relevant country names
at_use_df.to_csv('../Cluster_Analysis_Data/alcohol_tobacco-clean.csv', index=False)
```

```python
# Read in the adjusted content
at_df = pd.read_csv('../Cluster_Analysis_Data/alcohol_tobacco-clean.csv')
at_df = at_df.dropna()
at_df
```

```python
# Remove any countries that don't match the master list
at_clean_df = at_df.drop(at_df[at_df['country'].isin(no_match)].index.tolist())
at_clean_df
```

```python
# Write the CSV
at_clean_df.to_csv('../Cluster_Analysis_Data/alcohol_tobacco-clean.csv', index=False)
```
