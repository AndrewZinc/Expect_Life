```python
#  Import and read the updated input csv (After any manual updates to the CSV file)
pop_df = pd.read_csv('WPP2022_Population.csv')
pop_df
```

```python
# Drop the udpated file index column
pop_df = pop_df.drop(['LocTypeName', 'ParentID', 'VarID', 'Variant'], 1)
```

```python
pop_df
```

```python
pop_df = pop_df[pop_df['Time']>=2000]
pop_df = pop_df[pop_df['Time']<=2019]
pop_df
```

```python
pop_df.to_csv('WPP2022_Population.csv', index=False)
```

```python
# Open the master list of countries
countries_df = pd.read_csv('../master_country_list/country_profile_urls.csv')
countries_df = countries_df.drop(['profile_url'], 1)
```

```python
country_list = countries_df['country'].tolist()
country_list
```

```python
# Create a list of Food Supply countries
pop_countries = pop_df['Location'].tolist()
pop_countries
```

```python
# clean the FS data based on the SS country list
no_match = []

for country in pop_countries:
    if country in country_list:
        continue
    else:
        no_match.append(country)
```

```python
# Likely need to do manual country name clean-up for accented characters and abbreviations.
no_match
```

```python
pop_clean_df = pop_df.drop(pop_df[pop_df['Location'].isin(no_match)].index.tolist())
pop_clean_df
```

```python
pop_clean_df = pop_clean_df.dropna()
pop_clean_df
```

```python
pop_clean_df.to_csv('WPP2022_Population-clean.csv', encoding='utf-8', index=False)
```
