```python
# Import our dependencies
import pandas as pd
```

```python
#  Import and read the input csv
at_df = pd.read_csv('../Raw_Data/Cleaning_Completed/alcohol_tobacco_normalized.csv')
at_df
```

```python
# Open the master list of countries
countries_df = pd.read_csv('../Clean_Data/country_profile_urls.csv')
countries_df = countries_df.drop(['profile_url'], 1)
```

```python
country_list = countries_df['country'].tolist()
country_list
```

```python
# Create a list of life Expectancy countries
at_countries = at_df['Country Name'].tolist()
at_countries
```

```python
# clean the LE data based on the SS country list
no_match = []

for country in at_countries:
    if country in country_list:
        continue
    else:
        no_match.append(country)
```

```python
no_match
```

```python
at_clean_df = at_df.drop(at_df[at_df['Country Name'].isin(no_match)].index.tolist())
at_clean_df
```

```python
at_clean_df.to_csv('at_clean.csv', index=False)
```
