```python
# Import our dependencies
import pandas as pd
```

```python
#  Import and read the input csv
he_df = pd.read_csv("../Cluster_Analysis_Data/health_expenditure.csv")
he_df = he_df.dropna()
he_df
```

```python
govt_he = []
private_he =[]
govt_he_gdp =[]
```

```python
for i in he_df['Series Name'].index:
    if i % 3 == 0:
        govt_he.append(he_df['2019'][i])
govt_he
```

```python
for i in he_df['Series Name'].index:
    if i % 3 == 1:
        private_he.append(he_df['2019'][i])
private_he
```

```python
for i in he_df['Series Name'].index:
    if i % 3 == 2:
        govt_he_gdp.append(he_df['2019'][i])
govt_he_gdp
```

```python
he_df = he_df.drop_duplicates(subset=['Country Name'])
he_df
```

```python
# Create a list of health expenditure countries
he_countries = he_df['Country Name'].tolist()
he_countries
```

```python
# Build a new DataFrame from the component lists

health_ex_df = pd.DataFrame()
health_ex_df['country'] = he_countries
health_ex_df['govt_he'] = govt_he
health_ex_df['private_he'] = private_he
health_ex_df['govt_he_gdp'] = govt_he_gdp

health_ex_df
```

```python
# Open the master list of countries
countries_df = pd.read_csv('../master_country_list/country_profile_urls.csv')
#countries_df = countries_df.drop(['profile_url'], 1)
```

```python
country_list = countries_df['country'].tolist()
country_list
```

```python
# clean the HE data based on the SS country list
no_match = []

for country in he_countries:
    if country in country_list:
        continue
    else:
        no_match.append(country)
```

```python
# Likely need to do manual country name clean-up for accented characters and abbreviations.
no_match
```

```python
# Write the HE data to CSV to adjust relevant country names
health_ex_df.to_csv('../Cluster_Analysis_Data/health_expenditure-clean.csv', index=False)
```

```python
# Read in the adjusted content
he_df = pd.read_csv("../Cluster_Analysis_Data/health_expenditure-clean.csv")
he_df
```

```python
# Remove any countries that don't match the master list
he_clean_df = he_df.drop(he_df[he_df['country'].isin(no_match)].index.tolist())
he_clean_df
```

```python
# Write the CSV
he_clean_df.to_csv('../Cluster_Analysis_Data/health_expenditure-clean.csv', index=False)
```
