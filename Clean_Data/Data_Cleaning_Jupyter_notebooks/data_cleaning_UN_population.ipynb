```python
# Import our dependencies
import pandas as pd
```

```python
#  Import and read the input csv
pop_df = pd.read_csv("../Cluster_Analysis_Data/un_population_2022.csv")
#, encoding='iso-8859-1'
pop_df
```

```python
# Open the master list of countries
countries_df = pd.read_csv('../master_country_list/country_profile_urls.csv')
countries_df = countries_df.drop(['profile_url'], 1)
```

```python
country_list = countries_df['country'].tolist()
country_list
```

```python
# Create a list of life Expectancy countries
pop_countries = pop_df['country'].tolist()
pop_countries
```

```python
# clean the LE data based on the SS country list
no_match = []

for country in pop_countries:
    if country in country_list:
        continue
    else:
        no_match.append(country)
```

```python
# Likely need to do manual country name clean-up for accented characters and abbreviations.
no_match
```

```python
pop_clean_df = pop_df.drop(pop_df[pop_df['country'].isin(no_match)].index.tolist())
pop_clean_df
```

```python
pop_clean_df.to_csv('../Cluster_Analysis_Data/un_population_2022-clean.csv', index=False)
```
