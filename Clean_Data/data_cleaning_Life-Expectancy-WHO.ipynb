```python
# Import our dependencies
import pandas as pd
```

```python
#  Import and read the input csv
le_df = pd.read_csv("../Clean_Data/Cluster_Analysis_Data/WHO-LEx-data-separated.csv")
le_df
```

```python
# Open the master list of countries
countries_df = pd.read_csv('../Clean_Data/country_profile_urls.csv')
countries_df = countries_df.drop(['profile_url'], 1)
```

```python
country_list = countries_df['country'].tolist()
country_list
```

```python
# Create a list of life Expectancy countries
le_countries = le_df['country'].tolist()
le_countries
```

```python
# clean the LE data based on the SS country list
no_match = []

for country in le_countries:
    if country in country_list:
        continue
    else:
        no_match.append(country)
```

```python
# Likely need to do manual country name clean-up for accented characters and abbreviations.
no_match
```

```python
le_clean_df = le_df.drop(le_df[le_df['country'].isin(no_match)].index.tolist())
le_clean_df
```

```python
le_clean_df.to_csv('../Clean_Data/Cluster_Analysis_Data/WHO-LEx-clean.csv', index=False)
```
