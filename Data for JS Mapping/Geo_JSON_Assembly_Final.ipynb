```python
import pandas as pd

life_df = pd.read_csv("Global Life Expectancy by Country by Year - API_SP.DYN.LE00.IN_DS2_en_csv_v2_4700946.csv",)
life_df
```

```python
# Remove all columns besides country codes and expectancy data for latest available year
life_df = life_df[['Country Code', '2020']]
life_df
```

```python
import json

# Load existing geojson point data
with open('ne_10m_admin_0_label_points-country.json') as f:
  data = json.load(f)
data
```

```python
# for use with regex in order to add life expectancy data to geojson data
import re

for i in range(0,len(data['features'])):
  code = data['features'][i]['properties']['sr_adm0_a3']
  value = life_df.loc[life_df['Country Code'] == code, '2020']
  value1 = value.to_json()
  try:
    value2 = re.search("\d\d\.\d\d", value1).group(0)
    value3 = float(value2)
    data['features'][i]['properties']['life_expectancy'] = value3
  except:
    data['features'][i]['properties']['life_expectancy'] = 0
```

```python
# Write JSON file with new data
with open('point_data_with_life.json', 'w') as f:
  f.write(json.dumps(data))
```
